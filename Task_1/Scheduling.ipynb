```python
teachers = 5
lectures = 2
weeks = 2
days = 6
hours = 8
groups = 2

main_vars = teachers * lectures * weeks * days * hours * groups
cons_1 = teachers * weeks * days * hours
cons_2 = weeks * days * hours * groups
cons_3 = weeks * days * groups * 3
cons_4 = teachers * lectures * weeks * days * groups * 2

p = 50
BB = 10
B = 20
QUBO_size = main_vars + cons_1 + cons_2 + cons_3 + cons_4
QUBO = np.zeros((QUBO_size, QUBO_size))
QUBO.shape
```

## Constraint 1

```python
for teacher in range(teachers):
    for lecture in range(lectures):
        for week in range(weeks):
            for day in range(days):
                for hour in range(hours):
                    for group in range(groups):
                        index = teacher * lectures * weeks * days* hours * groups + \
                                week * days * hours * groups + \
                                day * hours * groups + \
                                hour * groups
                        
                        x = lecture * weeks * days * hours * groups + group
                        for l1 in range(lectures):
                            for g1 in range(groups):
                                y = l1 * weeks * days * hours * groups + g1
                                QUBO[index + x][index + y] += 1 * p

for teacher in range(teachers):
    for week in range(weeks):
        for day in range(days):
            for hour in range(hours):
                index = teacher * weeks * days * hours + \
                        week * days * hours + \
                        day * hours + \
                        hour
                QUBO[main_vars + index][main_vars + index] += 1 * p
                
for teacher in range(teachers):
    for lecture in range(lectures):
        for week in range(weeks):
            for day in range(days):
                for hour in range(hours):
                    for group in range(groups):
                        index = teacher * lectures * weeks * days* hours * groups + \
                                lecture * weeks * days * hours * groups + \
                                week * days * hours * groups + \
                                day * hours * groups + \
                                hour * groups + \
                                group
                        
                        y = teacher * weeks * days * hours + \
                            week * days * hours + \
                            day * hours + \
                            hour
                        
                        QUBO[index][main_vars + y] += 1 * p
                        QUBO[main_vars + y][index] += 1 * p
                        
for teacher in range(teachers):
    for lecture in range(lectures):
        for week in range(weeks):
            for day in range(days):
                for hour in range(hours):
                    for group in range(groups):
                        index = teacher * lectures * weeks * days * hours * groups + \
                                lecture * weeks * days * hours * groups + \
                                week * days * hours * groups + \
                                day * hours * groups + \
                                hour * groups + \
                                group
                        
                        QUBO[index][index] -= 2 * p

for teacher in range(teachers):
    for week in range(weeks):
        for day in range(days):
            for hour in range(hours):
                index = teacher * weeks * days * hours + \
                        week * days * hours + \
                        day * hours + \
                        hour
                QUBO[main_vars + index][main_vars + index] -= 2 * p
```
